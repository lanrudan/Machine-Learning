### PCA（主成分分析）

生成的矩阵无可读性

```python
'''
参数：
n_components：降维后我们需要的维度 如果希望可视化一组数据来观察数据分布，需要降维到三维以下
'''
```

```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
```

```python
iris = load_iris()
```

```python
x = iris.data
y = iris.target
```

```python
x.shape
```

```python
pca = PCA(n_components=2)
pca = pca.fit(x)
x_de = pca.transform(x)
x_de
```

```python
x_de[y == 0, 0]
```

```python
plt.figure()
plt.scatter(x_de[y == 0, 0], x_de[y == 0, 1], label = iris.target_names[0])
plt.scatter(x_de[y == 1, 0], x_de[y == 1, 1], label = iris.target_names[1])
plt.scatter(x_de[y == 2, 0], x_de[y == 2, 1], label = iris.target_names[2])
plt.legend()
plt.show()

# 适合用KNN
```

```python
pca.explained_variance_
# 可解释性方差的大小，即信息量大小
```

```python
pca.explained_variance_ratio_
# 查看降维后每个新特征向量所占的信息量站原始数据总信息量的百分比
# 可解释方差贡献率
```

```python
# 大部分信息集中在第一个特征上
# 特征砍掉一半，但是信息可以保留0.9777
pca.explained_variance_ratio_.sum()
```

```python
# 累计可解释方差贡献率曲线
# 选择最好的n_components
pca = PCA()
pca = pca.fit(x)
x_de = pca.transform(x)
x_de
```

```python
pca.explained_variance_
```

```python
import numpy as np
np.cumsum(pca.explained_variance_ratio_)
```

```python
# 核心解决中文乱码的代码
plt.rcParams['font.sans-serif'] = ['SimHei']  # 或者 'Microsoft YaHei'
plt.rcParams['axes.unicode_minus'] = False # 解决负号显示问题

plt.plot([1, 2, 3, 4], np.cumsum(pca.explained_variance_ratio_))
plt.xlabel("降维后特征数")
plt.ylabel("累积可解释性方差比率")
plt.show()
# 选转折点的特征数量
```

```python
# n_components输入mle（最大似然估计）作为超参数
# 计算量大
pca_mle = PCA(n_components='mle').fit(x)
x_mle = pca_mle.transform(x)
```

```python
# n_components输入信息量占比作为超参数
pca_f = PCA(n_components=0.97, svd_solver='full').fit(x)
x_f = pca_f.transform(x)
x_f
```

```python
pca_f.explained_variance_ratio_.sum()
```

### SVD（奇异值分解）

定义
对于任意 $m \times n$ 矩阵 $\mathbf{A}$，其**奇异值分解 (SVD)** 为：

$$
\mathbf{A} = \mathbf{U} \mathbf{\Sigma} \mathbf{V}^T
$$

其中：
- $\mathbf{U}$: $m \times m$ 正交矩阵 (左奇异向量)
- $\mathbf{\Sigma}$: $m \times n$ 对角矩阵 ($\mathbf{\Sigma}_{ii} = \sigma_i$)
- $\mathbf{V}^T$: $n \times n$ 正交矩阵的转置 (右奇异向量)


关键性质  
- **秩揭示**：非零 $\sigma_i$ 的数量 = $\text{rank}(A)$
- **最优近似**：最佳秩 $k$ 逼近：$A_k = \sum_{i=1}^k \sigma_i \mathbf{u}_i \mathbf{v}_i^T$
- **稳定性**：对噪声和数值扰动鲁棒  


几何意义  
SVD 将线性变换分解为三步：
1. 旋转/反射 ($V^T$)
2. 缩放 ($\Sigma$)
3. 旋转/反射 ($U$)




```python
PCA(2).fit(x).components_.shape # (k,n)
```

```python
'''
参数：
svd_sovler：控制矩阵分解细节的参数 
            可以选择：auto full arpack 
            **auto**：
            full：适合数据量适中，计算时间充足的情况
            arpark：可以加快运算速度，适合特征矩阵很大是时候，但一般用于特征矩阵为**稀疏矩阵**的情况
            **randomized**：比full快，还能保证效果。适合大矩阵
random_state
'''

'''
属性：
components_：新的特征矩阵
'''
```

### 用PCA做噪音过滤

先PCA降维，再inverse回原维

```python
from sklearn.datasets import load_digits
```

```python
digits = load_digits()
digits.data.shape
# 64个特征
```

```python
def plot_digits(data):
    fig, axes = plt.subplots(4, 10, figsize = (10, 4)
                             ,subplot_kw= {'xticks': [], 'yticks': []}
                             )
    for i, ax in enumerate(axes.flat):
        ax.imshow(data[i].reshape(8,8), cmap = 'binary')
```

```python
plot_digits(digits.data)
```

```python
# 由于数据集完美 我们认为加上噪音
np.random.RandomState(42)
noise = np.random.normal(digits.data, 2) # 向原数据集中的每个数据点添加独立的，均值为该数据点本身、标准差为2的高斯噪声
plot_digits(noise)
```

```python
# 开始降噪
pca = PCA(0.5, svd_solver='full').fit(noise)
x_de = pca.transform(noise)
```

```python
x_de.shape
# 仅有4个特征
```

```python
noise_inverse = pca.inverse_transform(x_de)
noise_inverse.shape
```

```python
plot_digits(noise_inverse)
```

### PCA对手写数字数据集的降维

```python
import pandas as pd
from sklearn.ensemble import RandomForestClassifier as RFC
from sklearn.model_selection import cross_val_score
data = pd.read_csv("train.csv")
```

```python
x = data.iloc[:, 1:]
y = data.iloc[:, 0]
```

```python
x.shape
```

```python
# 画累计方差贡献率曲线，找最佳降维后的维度范围
pca_line = PCA().fit(x)
plt.figure()
plt.plot(np.cumsum(pca_line.explained_variance_ratio_))
plt.show()
```

```python
# n_components选转折处0-200
# 画降维后的学习曲线
score = []
for i in range(1, 101, 10):
    x_de = PCA(i).fit_transform(x)
    defen = cross_val_score(RFC(n_estimators=10, random_state=0)
                            ,x_de
                            ,y
                            ,cv = 5
                ).mean()
    score.append(defen)
    # print(i)

plt.figure()
plt.plot(range(1, 101, 10), score)
plt.show()
```

```python
# 降维后的最佳维度为 20 左右
# 回忆特征选择的时候 当时是一半，即300多个特征 分数为0.96几几
score = []
for i in range(10, 25):
    x_de = PCA(i).fit_transform(x)
    defen = cross_val_score(RFC(n_estimators=10, random_state=0)
                            ,x_de
                            ,y
                            ,cv = 5
                ).mean()
    score.append(defen)

plt.figure()
plt.plot(range(10, 25), score)
plt.show()
```

```python
# 发现n_component取23时随机森林模型有最好的表现
x_de23 = PCA(23).fit_transform(x)
score_23 = cross_val_score(RFC(n_estimators=10, random_state=0)
                           ,x_de23
                           ,y
                           ,cv = 5).mean()
score_23
```

```python
# 调随机森林参数
score_23_100 = cross_val_score(RFC(n_estimators=100, random_state=0)
                           ,x_de23
                           ,y
                           ,cv = 5).mean()
score_23_100
```

```python
# 还是没有我们用嵌入法选择特征之后的结果高
# 选择KNN试试
from sklearn.neighbors import KNeighborsClassifier as KNN
```

```python
knn_score = cross_val_score(KNN(), x_de23, y, cv = 5).mean()
knn_score
```

```python
# KNN的k值学习曲线
score = []
for i in range(10):
    defen = cross_val_score(KNN(i + 1), x_de23, y, cv = 5).mean()
    score.append(defen)
#    print(i)

plt.figure()
plt.plot(range(10), score)
plt.show()
```

```python
# 当k = 4 + 1时模型效果最好
knn5 = cross_val_score(KNN(5), x_de23, y, cv = 5).mean()
knn5
# 交叉验证之后分数为
```
